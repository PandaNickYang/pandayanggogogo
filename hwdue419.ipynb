假設方程式 f(x)=x^3+5x^2+4x+3  其微分方程式為 f'(x)=3x^2+10x+4

```python
def f(x):
    return x**3+5*x**2+4*x+3
def df(x):
    return 3*x**2+10*x+4
```

```python
x = np.linspace(-7.7,8.8,100)
y = f(x)
plt.plot(x, y)         
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
plt.plot(x,y)
k=2
constant=0.004
for i in range(100):
    k=k-constant*df(k)
    plt.scatter(k,df(k))
```
